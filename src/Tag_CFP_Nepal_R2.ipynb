```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("CFP_Nepal_R2.csv",header=0, sep=';')
```

```python
df.loc[np.array(['short_term' in answer for answer in df['Q1.i']]), 'Q1.i'] = 'short_term_shelter__tent_shelt'
```

```python
df['District']=np.array([d.lower() for d in df['District']])
```

```python
count_list = []
for gender in ['female','male']:
    for district in df['District'].unique():
        for Q1answer in df['Q1.i'].unique():
            if Q1answer=='n/a':
                continue
            count = ((df['Gender']==gender) & (df['District']==district) & (df['Q1.i']==Q1answer)).sum()
            count_list.append({'District' : district, 'Gender' : gender, 'Q1.i' : Q1answer, 'Count' : count})
output_df = pd.DataFrame(count_list, columns=['District', 'Gender', 'Q1.i', 'Count'])          
```

```python
output_df.sort(['District', 'Gender', 'Count'], ascending = [1,1,0], inplace=True)
```

```python
# -- 
output_df.to_csv('rearranged_df.csv', index=False)
```

```python
# -- open 'rearranged_df.csv'
new_df = pd.read_csv('rearranged_df.csv')
```

```python
# -- determine 'Tag' from answer in Q1.i
def determine_tag(answer):
    if ('shelter' in answer) or ('housing' in answer) :
        return 'housing'
    elif 'water' in answer:
        return 'water'
    elif 'education' in answer:
        return 'education'
    elif 'financial' in answer:
        return 'financial'
    elif ('food' in answer) or ('seeds' in answer):
        return 'food'
    elif ('healthcare' in answer) or ('psycho' in answer) or ('sanitation' in answer):
        return 'health'
    # ...
    else:
        return None
```

```python
# -- add 'Tag' column
new_df['Tag'] = [determine_tag(tag) for tag in new_df['Q1.i']]
```
